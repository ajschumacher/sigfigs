```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(start=2, stop=7, num=1_000)
```

```python
y = scipy.stats.norm(loc=4.3, scale=0.25).pdf(x)
```

```python
fig1 = plt.figure(figsize=(13, 2))
plt.axvline(4.3, color='lightgrey')
plt.plot(x, y)
plt.yticks([])
plt.ylabel('probability')
plt.xlim([2.8, 6.2])
plt.xticks([3, 3.8, 4, 4.3, 4.8, 5, 6], [3, 3.8, 4, 4.3, 4.8, 5, 6])
plt.xlabel('quantity')
plt.title('With Significant Figures, “4.3” means that our belief about the true value of the quantity\n'
          'is described by a Gaussian probability distribution centered at 4.3, with standard deviation $\sigma = 0.25$,\n'
          'which means 95% confidence that the true value is between 3.8 and 4.8.')
plt.savefig('gaussian.png', dpi=160, bbox_inches='tight', transparent=False, facecolor='white')
```

```python
bad_y = np.where((4 <= x) & (x < 5), 1, 0)
```

```python
fig2 = plt.figure(figsize=(13, 2))
plt.axvline(4.3, color='lightgrey')
plt.plot(x, bad_y)
plt.yticks([])
plt.ylabel('probability')
plt.xlim([2.8, 6.2])
plt.xticks([3, 4, 4.3, 5, 6], [3, 4, 4.3, 5, 6])
plt.xlabel('quantity')
plt.title('If “4.3” is taken to mean the $\it{digit}$ 3 is uncertain, in the sense of '
          'a uniform probability distribution from 4 to 5,\n'
          'this implies we think 4.9 is as likely as 4.3, and 3.9 is impossible. '
          'This is probably not what we actually believe.')
plt.savefig('uniform.png', dpi=160, bbox_inches='tight', transparent=False, facecolor='white')
```
